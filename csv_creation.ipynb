```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
apple = yf.download("AAPL", start = "2000-01-01", end = "2024-12-31")
google = yf.download("GOOG", start = "2000-01-01", end = "2024-12-31")
apple.head()
```

```python
apple = pd.DataFrame(apple)
google = pd.DataFrame(google)
```

```python
apple = apple.reset_index()
google = google.reset_index()
```

```python
multi_index_a = pd.MultiIndex.from_tuples(apple.columns)
multi_index_g = pd.MultiIndex.from_tuples(google.columns)
```

```python
flat_columns_a = ['_'.join(col) for col in multi_index_a]
flat_columns_g = ['_'.join(col) for col in multi_index_g]

apple.columns = flat_columns_a
google.columns = flat_columns_g
```

```python
flat_columns_g
```

```python
df = pd.merge(apple, google, on = ['Date_', 'Date_'], how = 'outer')
```

```python
tickers = ["MSFT", "AMZN", "TSLA", "NVDA", "META", "JPM", "V", "BABA", "TSM",
    "SPY", "QQQ", "ARKK", "VOO", "DIA", "XLF",
    "BTC-USD", "ETH-USD", "DOGE-USD", "ADA-USD", "XRP-USD", "SOL-USD",
    "^GSPC", "^DJI"]

print(tickers)
```

```python
for ticker in tickers:
    stock_df = yf.download(ticker, start = "2000-01-01", end = "2024-12-31")

    stock_df = stock_df.reset_index()
    multi_index = pd.MultiIndex.from_tuples(stock_df.columns)
    flat_columns = ['_'.join(col) for col in multi_index]
    stock_df.columns = flat_columns
    
    df = pd.merge(df, stock_df, on = ['Date_', 'Date_'], how = 'outer')
```

```python
df
```

```python
df.to_csv("stocks.csv", index=True)
```
